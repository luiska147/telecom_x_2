## Feature encoding
## Codificación de features

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

```

```python
categorical_cols = categorical_cols.drop(['Genero', 'Tiene Servicio Telefonico'])
categorical_cols
```

```python
one_hot_encoder = make_column_transformer((OneHotEncoder(handle_unknown='ignore'), categorical_cols), remainder='passthrough')

one_hot_datos = one_hot_encoder.fit_transform(datos)
one_hot_datos = pd.DataFrame(one_hot_datos, columns= one_hot_encoder.get_feature_names_out())
one_hot_datos
```

## Separate data for training and testing
## Separar datos para entrenamiento y pruebas

```python
from sklearn.model_selection import train_test_split
X = one_hot_datos.drop(columns='remainder__Churn')
y = one_hot_datos['remainder__Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, stratify=y, random_state=20)
```

## Define baseline model
## Definir modelo base

```python
from sklearn.dummy import DummyClassifier

baseline = DummyClassifier(random_state=20)
baseline.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, f1_score, accuracy_score, precision_score, recall_score
from sklearn.metrics import classification_report
```

```python
y_pred_dummy = baseline.predict(X_test)

def calcular_regresion(y_test, y_pred):
  rmse = mean_squared_error(y_test, y_pred)
  mae = mean_absolute_error(y_test, y_pred)
  f1 = f1_score(y_test, y_pred, zero_division=0)
  acc = accuracy_score(y_test, y_pred)
  prec = precision_score(y_test, y_pred, zero_division=0)
  rec = recall_score(y_test, y_pred, zero_division=0)

  metricas = {
      'RMSE': round(rmse**(1/2), 4),
      'MAE': round(mae, 4),
      'F1': round(f1, 4),
      'Accuracy': round(acc, 4),
      'Precision': round(prec, 4),
      'Recall': round(rec, 4)
  }
  return metricas
```

```python
resultados_baseline = calcular_regresion(y_test, y_pred_dummy)
resultados_baseline
```

## Create Model Logistic Regression
## Crear modelo Regression Logistica

```python
from sklearn.linear_model import LogisticRegression

modelo_log = LogisticRegression(max_iter=1000, random_state=20)
modelo_log.fit(X_train, y_train)

y_pred_modelo_log = modelo_log.predict(X_test)
```

```python
resultados_modelo_log = calcular_regresion(y_test, y_pred_modelo_log)
resultados_modelo_log
```

## Create Random Forest Classifier model
## Crear modelo Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

modelo_rf = RandomForestClassifier(max_depth=5, random_state=20)
modelo_rf.fit(X_train, y_train)

y_pred_modelo_rf = modelo_rf.predict(X_test)

resultados_modelo_rf = calcular_regresion(y_test, y_pred_modelo_rf)
resultados_modelo_rf
```



# Informe Analítico: Predicción y Prevención de Churn (Fuga de Clientes) en Telecomunicaciones

## 1. Resumen Ejecutivo
El presente proyecto tiene como objetivo principal identificar a los clientes con alta probabilidad de cancelar sus servicios de telecomunicaciones (**Churn**) y entender los factores subyacentes que motivan esta decisión. Mediante la aplicación de técnicas de Machine Learning, se evaluaron múltiples modelos predictivos. El modelo final seleccionado (Regresión Logística Optimizada con penalización L1) logró identificar exitosamente al **80.45%** de los clientes dispuestos a abandonar la compañía, proporcionando una base sólida para implementar estrategias de retención proactivas.

---

## 2. Metodología y Preprocesamiento
Para asegurar la calidad de las predicciones, los datos crudos fueron sometidos a un riguroso pipeline de transformación:

* **Selección de Variables:** Se aplicó la prueba estadística de Chi-cuadrado ($\chi^2$) para descartar variables estadísticamente irrelevantes (ej. Género, Servicio Telefónico), reduciendo el ruido dimensional.
* **Codificación (Encoding):** Transformación de variables categóricas múltiples mediante *One-Hot Encoding* para hacerlas interpretables por los algoritmos.
* **Escalamiento:** Normalización de variables numéricas continuas para asegurar que características con rangos amplios (como `Cargos Totales`) no dominaran el peso del modelo.

---

## 3. Análisis Crítico y Comparación de Modelos
Durante la fase de experimentación, se contrastaron principalmente modelos basados en ensambles de árboles (Random Forest) y modelos estadísticos lineales (Regresión Logística).

### Desempeño General
El mejor desempeño, alineado con los objetivos del negocio, lo obtuvo la **Regresión Logística optimizada mediante GridSearchCV** (`C=1`, `penalty='l1'`, optimizada por la métrica `roc_auc`).

Aunque los modelos iniciales de Random Forest lograron una mayor exactitud global (Accuracy > 80%), fallaban en la métrica más crítica: el **Recall**. Solo lograban detectar a la mitad de los clientes en riesgo (~50%). La Regresión Logística optimizada sacrificó un poco de exactitud general (bajando a 75%), pero maximizó el **Recall a un 80.45%**, logrando además el mejor equilibrio con un **F1-Score de 0.6199**.

### Análisis de Overfitting y Underfitting
* **Modelos Iniciales (Overfitting):** Los primeros intentos tendían a memorizar la clase mayoritaria (clientes que se quedan), mostrando alta precisión pero baja capacidad para generalizar la clase minoritaria (Churn). Esto es un claro síntoma de sobreajuste hacia la clase dominante.
* **Modelo Final Equilibrado:** Al introducir la penalización `L1` (Lasso) en la Regresión Logística, el modelo forzó a cero los coeficientes de las variables menos relevantes. Esto actuó como un filtro contra el sobreajuste. El modelo resultante es lo suficientemente complejo para capturar la tendencia de fuga, sin memorizar el ruido de los datos de entrenamiento.

---

## 4. Análisis de la Matriz de Confusión
Basado en las predicciones del modelo final sobre el conjunto de prueba (Test Set), el comportamiento del algoritmo fue el siguiente:

| Predicción \ Realidad | Cliente se Queda (No Churn) | Cliente se Va (Churn) |
| :--- | :--- | :--- |
| **Predice: No Churn** | **961** (Verdaderos Negativos) | **87** (Falsos Negativos) |
| **Predice: Churn** | **352** (Falsos Positivos) | **358** (Verdaderos Positivos) |

**Interpretación de Negocio:**
* **El Éxito (358 VP):** Clientes que se iban a ir y el modelo detectó a tiempo.
* **El Punto Ciego (87 FN):** Clientes que se fueron sin ser detectados. Al ser un número tan bajo (solo el 19.5% de los desertores totales), confirma el éxito del alto Recall del modelo.
* **El Costo de Oportunidad (352 FP):** Clientes leales que el modelo marcó como riesgo. En el contexto de retención, contactar a estos clientes para ofrecerles una mejora no supone una pérdida grave e incluso fomenta la fidelización (*upselling/cross-selling*).

---

## 5. Factores Clave de Cancelación (Feature Importance)
El análisis de los coeficientes extraídos del modelo final reveló los siguientes impulsores de retención y cancelación:

**Factores de Alto Riesgo (Aumentan la probabilidad de Churn):**
1. **Contrato Mes a Mes:** Es el factor más crítico. La falta de compromiso a largo plazo facilita la migración rápida hacia la competencia.
2. **Servicio de Fibra Óptica:** Los usuarios con este servicio presentan mayores tasas de cancelación. Esto suele indicar expectativas no cumplidas (fallas de servicio) o precios poco competitivos en ese segmento específico.
3. **Cargos Mensuales Elevados:** A mayor costo recurrente, mayor sensibilidad a ofertas externas.

**Factores de Retención (Disminuyen la probabilidad de Churn):**
1. **Contratos a Dos Años:** Generan alta fidelidad y sólidas barreras de salida.
2. **Ausencia de Servicio de Internet:** Los clientes que solo consumen servicios básicos o tradicionales son los más estables del portafolio.

---

## 6. Estrategias de Retención Propuestas
Con base en los hallazgos respaldados por el modelo, se recomiendan las siguientes acciones estratégicas:

1. **Campaña de Migración de Contratos:** Diseñar ofertas atractivas (ej. descuentos en los primeros meses o mejoras de velocidad sin costo) dirigidas exclusivamente a los clientes con "Contratos Mes a Mes", condicionadas a la firma de un contrato de fidelidad de 1 o 2 años.
2. **Auditoría de Calidad en Fibra Óptica:** Iniciar una investigación técnica y de servicio al cliente enfocada en los usuarios de Fibra Óptica para identificar si las cancelaciones se deben a fallas en la red en regiones específicas o inconformidad con el precio.
3. **Focalización del Presupuesto de Retención:** Utilizar las probabilidades generadas por este algoritmo para dirigir los esfuerzos del equipo de *Call Center* exclusivamente a los clientes marcados como alto riesgo. Esto optimizará drásticamente el presupuesto, evitando ofrecer descuentos masivos a clientes que no tenían intención de abandonar el servicio.

# Analytical Report: Churn Prediction and Prevention in Telecommunications

## 1. Executive Summary
The primary objective of this project is to identify customers with a high probability of canceling their telecommunications services (**Churn**) and to understand the underlying factors driving this decision. Through the application of Machine Learning techniques, multiple predictive models were evaluated. The final selected model (Optimized Logistic Regression with L1 penalty) successfully identified **80.45%** of the customers willing to leave the company, providing a solid foundation for implementing proactive retention strategies.

---

## 2. Methodology and Preprocessing
To ensure the quality of predictions, the raw data underwent a rigorous transformation pipeline:

* **Feature Selection:** The Chi-square ($\chi^2$) statistical test was applied to discard statistically irrelevant variables (e.g., Gender, Phone Service), reducing dimensional noise.
* **Encoding:** Transformation of multiple categorical variables using *One-Hot Encoding* to make them interpretable by the algorithms.
* **Scaling:** Normalization of continuous numerical variables to ensure that features with wide ranges (such as `Total Charges`) did not dominate the model's weights.

---

## 3. Critical Analysis and Model Comparison
During the experimentation phase, tree-based ensemble models (Random Forest) and linear statistical models (Logistic Regression) were primarily contrasted.

### Overall Performance
The best performance, aligned with business objectives, was achieved by the **Logistic Regression optimized via GridSearchCV** (`C=1`, `penalty='l1'`, optimized by the `roc_auc` metric).

Although the initial Random Forest models achieved higher overall accuracy (Accuracy > 80%), they failed in the most critical metric: **Recall**. They only managed to detect half of the at-risk customers (~50%). The optimized Logistic Regression sacrificed some overall accuracy (dropping to 75%) but maximized **Recall to 80.45%**, also achieving the best balance with an **F1-Score of 0.6199**.

### Overfitting and Underfitting Analysis
* **Initial Models (Overfitting):** The first attempts tended to memorize the majority class (retained customers), showing high precision but low ability to generalize the minority class (Churn). This is a clear symptom of overfitting toward the dominant class.
* **Balanced Final Model:** By introducing the `L1` penalty (Lasso) in the Logistic Regression, the model forced the coefficients of less relevant variables to zero. This acted as a filter against overfitting. The resulting model is complex enough to capture the churn trend without memorizing the noise of the training data.

---

## 4. Confusion Matrix Analysis
Based on the final model's predictions on the Test Set, the algorithm's behavior was as follows:

| Prediction \ Reality | Customer Stays (No Churn) | Customer Leaves (Churn) |
| :--- | :--- | :--- |
| **Predicts: No Churn** | **961** (True Negatives) | **87** (False Negatives) |
| **Predicts: Churn** | **352** (False Positives) | **358** (True Positives) |

**Business Interpretation:**
* **The Success (358 TP):** Customers who were going to leave and the model detected in time.
* **The Blind Spot (87 FN):** Customers who left without being detected. Being such a low number (only 19.5% of total churners), it confirms the success of the model's high Recall.
* **The Opportunity Cost (352 FP):** Loyal customers that the model flagged as at-risk. In the context of retention, contacting these customers to offer them an upgrade does not represent a severe loss and even fosters loyalty (*upselling/cross-selling*).

---

## 5. Key Churn Factors (Feature Importance)
The analysis of the coefficients extracted from the final model revealed the following retention and churn drivers:

**High-Risk Factors (Increase Churn probability):**
1. **Month-to-Month Contract:** This is the most critical factor. The lack of long-term commitment facilitates rapid migration to competitors.
2. **Fiber Optic Service:** Users with this service show higher cancellation rates. This usually indicates unmet expectations (service failures) or uncompetitive pricing in that specific segment.
3. **High Monthly Charges:** The higher the recurring cost, the greater the sensitivity to external offers.

**Retention Factors (Decrease Churn probability):**
1. **Two-Year Contracts:** Generate high loyalty and solid exit barriers.
2. **No Internet Service:** Customers who only consume basic or traditional services are the most stable in the portfolio.

---

## 6. Proposed Retention Strategies
Based on the findings supported by the model, the following strategic actions are recommended:

1. **Contract Migration Campaign:** Design attractive offers (e.g., discounts in the first months or speed upgrades at no cost) targeted exclusively at customers with "Month-to-Month Contracts," conditional on signing a 1 or 2-year loyalty contract.
2. **Fiber Optic Quality Audit:** Initiate a technical and customer service investigation focused on Fiber Optic users to identify whether cancellations are due to network failures in specific regions or dissatisfaction with the price.
3. **Targeting Retention Budget:** Use the probabilities generated by this algorithm to direct the *Call Center* team's efforts exclusively to customers flagged as high risk. This will drastically optimize the budget, avoiding mass discounts to customers who had no intention of leaving the service.
